```python
import pandas as  pd
import numpy as  np
import matplotlib.pyplot as  plt
from sklearn import linear_model
import joblib
```

```python
df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df10=df.head(10)
df10
```

```python
plt.xlabel("Area")
plt.ylabel("Price")

plt.scatter(df10.total_rooms, df10.median_house_value, color="red", marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df10[['total_rooms']], df10.median_house_value)
```

```python
plt.xlabel("Area")
plt.ylabel("Price")

plt.scatter(df10.total_rooms, df10.median_house_value, color="red", marker="+")
plt.plot(df10.total_rooms, reg.predict(df10[['total_rooms']]),color='blue')
```

```python
reg.predict([[6000]])
```

### **CONCEPT**

REGRESSION y= mx+c

m=coeficient

c=intercept

```python
reg.coef_
```

```python
reg.intercept_
```
